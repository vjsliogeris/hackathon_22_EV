```python
import fastbook
fastbook.setup_book()
from fastbook import *
from fastai.vision.widgets import *
from pathlib import Path
```

```python
path = Path("/home/eimantas/Desktop/HACKHATON/TASK2/images")
items = DataBlock(
    blocks     = (ImageBlock, CategoryBlock),
    get_items  = get_image_files,
    splitter   = RandomSplitter(valid_pct=0.2, seed=24),
    get_y      = parent_label,
    item_tfms  = Resize(256)
)
```

```python
items = items.new(
    item_tfms=RandomResizedCrop(256, min_scale=0.3))
dls = items.dataloaders(path, bs=32)
```

```python
dls.train.show_batch(max_n=5)
```

```python
learn = vision_learner(dls, resnet18, metrics=error_rate)
```

```python
learn.fine_tune(1)
```

```python
learn.fine_tune(2)
```
